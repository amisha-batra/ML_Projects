K Means Clustering 

```python
import matplotlib.pyplot as plt 
import pandas as pd 
from sklearn.datasets import make_blobs 
import numpy as np 
import seaborn as s 
%matplotlib inline 
```

```python
x,y = make_blobs(n_samples = 1000,centers = 3,n_features = 2)
```

```python
plt.scatter(x[:,0],x[:,1],c=y)
```

```python
## standidization feature scalling technique 
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.33,random_state=42)
```

```python
x_train_S =scaler.fit_transform(x_train)
x_test_S = scaler.fit(x_test)
```

```python
from sklearn.cluster import KMeans

## Elbow method 
wcss = []
for k in range(1,20):
    kmeans = KMeans(n_clusters=k,init='k-means++')
    kmeans.fit(x_train_S)
    wcss.append(kmeans.inertia_)

wcss
```

```python
## plot the elbow curve 
plt.plot(range(1,20),wcss)
plt.xticks(range(1,20))
```

```python
kmeans = KMeans(n_clusters = 3,init='k-means++')
kmeans.fit_transform(x_train_S)
```

```python
y = kmeans.predict(x_train_S)
y
```

```python
plt.scatter(x_test[:, 0], x_test[:, 1], c=y_test, cmap='plasma')
```

```python
from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans

sil = []

# Start from 2 because silhouette_score requires at least 2 clusters
for k in range(2, 20):
    kmeans = KMeans(n_clusters=k, init='k-means++', random_state=42)
    kmeans.fit(x_train_S)
    score = silhouette_score(x_train_S, kmeans.labels_)
    sil.append(score)

# Optional: Plot the silhouette scores
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
plt.plot(range(2, 20), sil, marker='o')
plt.xticks(range(2, 20))
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Silhouette Score")
plt.title("Silhouette Score vs. Number of Clusters")
plt.grid(True)
plt.show()
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

sil = []
models = []

# Step 1: Try K from 2 to 19
for k in range(2, 20):
    kmeans = KMeans(n_clusters=k, init='k-means++', random_state=42)
    kmeans.fit(x_train_S)
    score = silhouette_score(x_train_S, kmeans.labels_)
    sil.append(score)
    models.append(kmeans)

# Step 2: Find the best k
best_k_index = sil.index(max(sil))
best_k = best_k_index + 2  # because range started at 2
best_model = models[best_k_index]

# Step 3: Final chosen centroids
final_centroids = best_model.cluster_centers_
print(f"Best number of clusters: {best_k}")
print("Final chosen centroids:\n", final_centroids)

# Step 4: Optional  Plot silhouette scores
plt.figure(figsize=(8, 5))
plt.plot(range(2, 20), sil, marker='o')
plt.xticks(range(2, 20))
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Silhouette Score")
plt.title("Silhouette Score vs. Number of Clusters")
plt.grid(True)
plt.show()
```
